# How matplotlib figuare we reate full method here

```python
#1. first import matplot lib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.image import imread

#2. prepare data
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 33, 44, 55,66]

#3. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

#4. Plot data
ax.plot(x,y, color="pink")

#5. Customize Plots
ax.set(title="Heart Dicease",
      xlabel="x-axis",
      ylabel="y-axis")

#6. Save and show, you save the whole figure
fig.savefig("images/heart-dicease.png")
```

# Creating matplot lib plot with numpy arrays

```python
x = np.linspace(0, 10, 100)
x
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, color="yellow")
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# bar plot
nut_prices = {
    "peanut": 12,
    "butter": 18,
    "Ghee": 90
}
fig, ax = plt.subplots()
ax.bar(nut_prices.keys(), nut_prices.values())
ax.set(title="Peanut butters prices", ylabel="Prices $");
```

```python
#bar horizontal
fig, ax = plt.subplots()
ax.barh(list(nut_prices.keys()), list(nut_prices.values()));
```

```python
#histogram
rand = np.random.randn(1000)
rand
fig, ax = plt.subplots()
ax.hist(rand);
```

```python
### subplot option 1
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_prices.keys(), nut_prices.values())
ax4.hist(np.random.random(1000));
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_prices.keys(), nut_prices.values())
ax[1, 1].hist(np.random.random(1000));
```

# Which one we should use when plotting
when we have to plot quick we can use pyplot 
other wise we can use 00 method

```python
heart_discease = pd.read_csv("heart-disease.csv")
```

```python
heart_discease.head()
```

```python
over_50 = heart_discease[heart_discease['age'] > 50]
```

```python
over_50
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# 00 method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45, 100])
```
